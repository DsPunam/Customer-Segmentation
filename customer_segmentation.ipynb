```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
x=df.iloc[:,[3,4]].values
x
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=40)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("number of clustering point")
plt.ylabel("wcss")
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=0)
y=kmeans.fit_predict(x)
y
```

```python
df["cluster"]=y
df.head()
```

```python
df0=df[df.cluster==0]
df0
```

```python
df1=df[df.cluster==1]
df1
```

```python
df2=df[df.cluster==2]
df2
```

```python
df3=df[df.cluster==3]
df3
```

```python
df4=df[df.cluster==4]
df4
```

```python
kmeans.cluster_centers_
```

```python

plt.scatter(df0["Annual Income (k$)"],df0["Spending Score (1-100)"],color="green",s=50)
plt.scatter(df1["Annual Income (k$)"],df1["Spending Score (1-100)"],color="red",s=50)
plt.scatter(df2["Annual Income (k$)"],df2["Spending Score (1-100)"],color="blue",s=50)
plt.scatter(df3["Annual Income (k$)"],df3["Spending Score (1-100)"],color="pink",s=50)
plt.scatter(df4["Annual Income (k$)"],df4["Spending Score (1-100)"],color="yellow",s=50)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black",s=100)
plt.xlabel("anual income")
plt.ylabel("income")
```
